### Understanding and Predicting Airbnb Prices in NYC

Using [Airbnb data](http://insideairbnb.com/get-the-data.html) scraped on June 2, 2021, our goal is to 
1.  Understand how price varies with factors such as geographic location, ratings, property type
2.  Build a model to predict listing price based on those relevant features

### Part 2: Price Prediction

Based on our initial exploration, it seems a little tricky to predict the precise price of an Airbnb listing given that there is so much variation in price at every value or category. Nonetheless, let's see how well some of these covariates can explain pricing by building some linear regression models.   

```python
cat_vars = ['neighbourhood_group_cleansed', 'room_type', 'instant_bookable'] 
cat_cols = pd.get_dummies(df[cat_vars], prefix=cat_vars, dummy_na=False, drop_first=True)
num_vars = ['beds', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value', 'price']
df1 = pd.concat([cat_cols, df[num_vars]], axis=1)
```

```python
# exclude missing values (review scores)
df1 = df1.dropna()
df1.shape
```

```python
# Simple linear regression model with covariates defined above and price as the outcome variable
X = df1.drop('price', axis=1)
y = df1['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=36)

lm = LinearRegression(normalize=True)
lm.fit(X_train, y_train)

y_test_preds = lm.predict(X_test)
y_train_preds = lm.predict(X_train)

mse_test = metrics.mean_squared_error(y_test, y_test_preds)
mse_train = metrics.mean_squared_error(y_train, y_train_preds)

r2_test = metrics.r2_score(y_test, y_test_preds)
r2_train = metrics.r2_score(y_train, y_train_preds)

print('Test error: {} Train error: {}'.format(mse_test, mse_train))
print('Test R-squared: {} Train R-squared: {}'.format(r2_test, r2_train))
```

The model does not fit the data very well -- mean squared error is huge, and R-squared is pretty low. Test error and R squared is slightly better than train, but that could be due to randomness in how we split the data. 

Let's take a look at the coefficients.

```python
def sort_coef(coefficients, coef_names):
    """
    Returns a list of coefficients sorted by absolute values 
    
    Args:
        coefficients (array): list of coefficients of size (n_features,)
        coef_names (list): list of corresponding coefficient names
    """
    coef = pd.DataFrame()
    coef['feature'] = coef_names
    coef['coefficient'] = np.round(coefficients, 4)
    coef['abs_coefficient'] = np.abs(coef['coefficient'])
    coef = coef.sort_values('abs_coefficient', ascending=False)
    return coef
```

```python
sort_coef(lm.coef_, X.columns)
```

The direction of a lot of the coefficients correspond to what we saw in the plots earlier. On average, a shared room or private room is cheaper than an entire home, whereas a hotel room tend to be more expensive. Location is an important predictor of price, as seen in the size of the coefficient for Manhattan and location review score. Number of beds as an indicator of the size of the space is also a salient factor.

Next, let's try adding in some interactions for the boroughs since there might be differences in how price varies with different attributes by borough. 

```python
poly = PolynomialFeatures(interaction_only=True, include_bias=False)
X_interact = poly.fit_transform(X)
X_interact.shape
```

```python
def fit_linear(X, y, test_size=0.3, random_state=42, normalize=True):
    """
    Splits data in train and test sets, fits a linear regression model and return results
    
    Args:
        X (dataframe): covariate matrix of size (n_features, n_samples)
        y (array): list of response variable values (n_samples,)
        test_size (float): fraction of data to withhold as test (default=0.3)
        random_state: random state for splitting train and test
        normalize (bool): if True, normalize the data first before fitting the linear regression model 
        
    Returns: 
        A dictionary of items and their values including: mean squared error for test and train,
        R-squared for test and train, fitted model, test predictions, train predictions, 
        X_train, X_test, y_train, y_test
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    lm = LinearRegression(normalize=normalize)
    lm.fit(X_train, y_train)

    y_test_preds = lm.predict(X_test)
    y_train_preds = lm.predict(X_train)
    
    mse_test = metrics.mean_squared_error(y_test, y_test_preds)
    mse_train = metrics.mean_squared_error(y_train, y_train_preds)
    r2_test = metrics.r2_score(y_test, y_test_preds)
    r2_train = metrics.r2_score(y_train, y_train_preds)
    names = ['mse_test', 'mse_train', 'r2_test', 'r2_train', 'model', 'y_test_preds', 'y_train_preds', 'X_train', 'X_test', 'y_train', 'y_test']
    results = dict(zip(names, [mse_test, mse_train, r2_test, r2_train, lm, y_test_preds, y_train_preds, X_train, X_test, y_train, y_test]))
    return results
```

```python
y = df1.price
results = fit_linear(X_interact, y, random_state=36)
print('Test mean squared error: {} Train error: {}'.format(results['mse_test'], results['mse_train']))
print('Test R squared: {} Train R squared: {}'.format(results['r2_test'], results['r2_train']))
```

```python
sort_coef(results['model'].coef_, poly.get_feature_names(list(X.columns)))[:20]
```

R2 improved only marginally compared to the model without interaction terms.

A lot of terms were added but they led little to no improvement in performance. Let's apply regularization to reduce the number of features and do cross validation to get a more reliable measure of model performance.

```python
ridge_cv = RidgeCV().fit(results['X_train'], results['y_train'])
print('R2 for test:', ridge_cv.score(results['X_test'], results['y_test']))
print('R2 for train:', ridge_cv.score(results['X_train'], results['y_train']))
```

Adding regularization helped improve test performance, but not by a lot.

After trying different things based on the features we initially selected, the fit is still poor. Are there important features that we're missing? Even though there wasn't a clear visual relationship, these following variables might still be useful:
- number of bedrooms (bigger property should be more expensive)
- availability in the future (more popular properties might be more expensive)
- host response rate

Let's add them one by one.

Adding the number of bedrooms improved R squared quite a bit!

Looking back at the plots earlier, let's add square terms for availability, number of reviews, host response rate, and review scores.

```python
num_vars = ['bedrooms', 'availability_30', 'number_of_reviews', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value', 'price']
df3 = pd.concat([cat_cols, df[num_vars]], axis=1)
df3 = df3.dropna()
sq_vars = ['availability_30', 'number_of_reviews', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value']
poly = PolynomialFeatures(interaction_only=True, include_bias=False)
sq_cols = poly.fit_transform(df3[sq_vars])
```

```python
def square_columns(df, colnames):
    """
    Creates squared columns from the list of columns given and append it to the existing dataframe
    
    Args:
        df: pandas dataframe
        colnames: list of column names to square
    
    Returns:
        a new dataframe with the squared columns appended
    """
    sq_cols = df[colnames].pow(2).add_prefix('squared_')
    df = pd.concat([df, sq_cols], axis=1)
    return df
```

```python
num_vars = ['bedrooms', 'beds', 'availability_30', 'host_response_rate', 'number_of_reviews', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value', 'price']
df3 = pd.concat([cat_cols, df[num_vars]], axis=1)
df3 = df3.dropna()

sq_vars = ['availability_30','number_of_reviews', 'host_response_rate', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location', 'review_scores_value']
df3 = square_columns(df3, sq_vars)

X = df3.drop('price', axis=1)
y = df3['price']

results = fit_linear(X, y)
print('Test mean squared error: {} Train error: {}'.format(results['mse_test'], results['mse_train']))
print('Test R squared: {} Train R squared: {}'.format(results['r2_test'], results['r2_train']))
```

```python
sort_coef(results['model'].coef_, X.columns)
```

```python
plt.scatter(results['y_test_preds'], results['y_test']-results['y_test_preds'], alpha=0.6, color='turquoise')
```

The errors still appear to be correlated with the outcome - the larger the value, the bigger the absolute error. Instead of OLS, Generalized least squares (GLS) may be more suitable for data that exibit heteroscedasticity.  

In conclusion, location (as measured by borough, location rating), room type, number of bedrooms, and cleanliness are the most salient features in predicting price.

There are, of course, lots of other factors that could affect price that we did not look at. For example, these might include:
* review recency 
* keywords in property description
* number of properties in the same neighborhood (competition/supply)
